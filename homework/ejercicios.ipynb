```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#
# Crear carpeta de salida si no existe
#

os.makedirs("C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/docs", exist_ok=True)
```

```python
#
# Carga de datos
#
def load_data():
    df = pd.read_csv('C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/input/shipping-data.csv')
    return df
```

```python
#
# Visual for Shipping per Warehouse
#
def create_visual_for_shipping_per_warehouse(df):
    df = df.copy()
    plt.figure()
    counts = df.Warehouse_block.value_counts()
    counts.plot.bar(
        title='Shipping per Warehouse',
        xlabel='Warehouse block',
        ylabel='Record Count',
        color='tab:blue',
        fontsize=8,
    )
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)
    plt.savefig("C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/docs/shipping_per_warehouse.png")
```

```python
#
# Visual for Mode of Shopment
#
def create_visual_for_mode_of_shipment(df):
    df = df.copy()
    plt.figure()
    counts = df.Mode_of_Shipment.value_counts()
    counts.plot.pie(
        title="Mode of Shipment",
        wedgeprops=dict(width=0.35),
        ylabel="",
        colors=["tab:blue", "tab:orange", "tab:green"],
    )
    plt.savefig("C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/docs/mode_of_shipment.png")

df = load_data()
create_visual_for_mode_of_shipment(df)
```

```python
#
# Visual for Average Customer Rating
#
def create_visual_for_average_customer_rating(df):
    df = df.copy()
    plt.figure()
    df = (
        df[["Mode_of_Shipment", "Customer_rating"]]
        .groupby("Mode_of_Shipment")
        .describe()
    )
    df.columns = df.columns.droplevel(0)
    df = df[["mean", "min", "max"]]
    plt.barh(
        y=df.index.values,
        width=df["max"].values - 1,
        left=df["min"].values,
        height=0.9,
        color="lightgray",
        alpha=0.8,
    )
    colors = [
        "tab:green" if value >=3.0 else "tab:orange" for value in df["mean"].values
    ]
    plt.barh(
        y=df.index.values,
        width=df["mean"].values - 1,
        left=df["min"].values,
        color=colors,
        height=0.5,
        alpha=1.0,
    )
    plt.title("Average Customer Rating")
    plt.gca().spines["left"].set_color("gray")
    plt.gca().spines["bottom"].set_color("gray")
    plt.gca().spines["top"].set_visible(False)
    plt.gca().spines["right"].set_visible(False)
    plt.savefig("C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/docs/average_customer_rating.png")
```

```python
#
# Visual for Weight Distribution
#
def create_visual_for_weight_distribution(df):
   df = df.copy()
   plt.figure()
   df.Weight_in_gms.plot.hist(
      title="Shipped Weight Distribution",
      color="tab:orange",
      edgecolor="white",
   ) 
   plt.gca().spines["top"].set_visible(False)
   plt.gca().spines["right"].set_visible(False)
   plt.savefig("C:/Github/Analitica_Descriptiva/Labs/LAB-08-matplotlib-dashboard-JuanTA97/files/docs/weight_distribution.png")
```

```python
df = load_data()
create_visual_for_shipping_per_warehouse(df)
create_visual_for_mode_of_shipment(df)
create_visual_for_average_customer_rating(df)
create_visual_for_weight_distribution(df)
```
